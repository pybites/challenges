# 100 Days of Random Awesome Python!!
This creates a curriculum of 100 days of random python awesomeness. Each day you get a new library to check out and explore.

```python
import random
import json

import requests
import parse
```

## Title, version, and repo url

```python
title = '100 Days of Awesome Python'
version = 0.1
github_repo = 'https://github.com/pybites/100DaysOfCode'
```

## Grab list of awesome python - curated list of awesome Python projects

```python
r = requests.get('https://raw.githubusercontent.com/vinta/awesome-python/master/README.md')
```

```python
lib = parse.compile('* [{lib_name}]({url}) - {description}')
```

```python
libs = [lib.parse(line).named for line in r.text.split('\n') if lib.parse(line)]
```

```python
print(f'Found {len(libs)} awesome projects to explore!')
```

## Create random list for tasks

```python
def get_random_library(libs):
    lib = libs.pop(random.randrange(len(libs)))
    return f'Explore Awesome library {lib["lib_name"]} - {lib["description"]} ({lib["url"]})'
```

```python
tasks = [{'day': day,
         'activity': get_random_library(libs),
         'done': False}
        for day in range(1, 101)]
```

## Generate JSON response

```python
result = json.dumps({'title': title,
                    'version': version,
                    'github_repo': github_repo,
                    'tasks': tasks}, 
                    sort_keys=False,
                    indent=1)
```

```python
print(result)
```
